```python
import random

def popravljanje (graf):
    index = 0
    for povezava in graf:
        index += 1
        (u, v, c) = povezava
        print('povezava {}'.format((u, v, c)))
        counter_u = 0
        counter_v = 0
        for ostale in graf:
            (u1, v1, c1) = ostale
            if (u1, v1, c1) == (u, v, c):
                counter_u += 1
                counter_v += 1
            elif (u == u1 or u == v1) and (v != u1 or v != v1):
                if c1 == c:
                    counter_u += 1
            elif (v == u1 or v == v1) and (u != u1 or u != v1):
                if c1 == c:
                    counter_v += 1
            print(counter_u, counter_v)
        if (counter_u, counter_v) == (1, 1):
            print(graf)
            for i, ostale in enumerate(graf):
                print(i, ostale)
                (u1, v1, c1) = ostale
                if (u == u1 or u == v1) and (povezava != ostale):
                    graf[i] = (u1, v1, c)
                    print graf
                    popravljanje(graf)
                else: pass
        elif counter_u == counter_v:              #pogoj za (2,2) in (3,3)
            print(graf)
            for i, ostale in enumerate(graf):
                print(i, ostale)
                (u1, v1, c1) = ostale
                if (u == u1 or u == v1) and (povezava != ostale) and (c == c1):
                    barva = [1, 2, 3]
                    barva.remove(c)
                    barva = random.choice(barva)
                    graf[i] = (u1, v1, barva)
                    print graf
                    popravljanje(graf)
                else: pass
        else: pass
    return graf
```
